# 종목 유사도 구하기

```python
df = pd.read_csv(f"/content/drive/MyDrive/Fintshift/Data&Code/ForViz/유사도 추천/추천_temp.csv")
```

```python
df1 = df[(df.iloc[:, 1:] != 0).any(axis=1)].reset_index(drop = True)
df2 = df1.set_index('종목명')
print(df2.shape)
df2.head(2)
```

```python
from scipy.spatial.distance import cosine
similarities_data = []

for index, row in df2.iterrows():
    distances = df2.apply(lambda x: cosine(row, x), axis=1)
    similarity_scores = 1 - distances  # 코사인 유사도는 1에서 거리를 뺀 값
    most_similar = similarity_scores.sort_values(ascending=False).index[1:6].tolist()
    similarities_data.append([index] + most_similar)

# 새로운 데이터프레임 생성
columns = ['종목명'] + [f'유사종목_{i}' for i in range(1, 6)]
similarities_df = pd.DataFrame(similarities_data, columns=columns)

similarities_df
```

```python
df1[(df1["종목명"]== "동화약품") | (df1["종목명"] == "빛샘전자")| (df1["종목명"] == "화일약품")]
```

```python
similarities_df1 = similarities_df.melt(id_vars = "종목명",
                                        value_vars = ["유사종목_1", "유사종목_2", "유사종목_3", "유사종목_4", "유사종목_5"])
similarities_df1 = similarities_df1.sort_values(["종목명", "variable"]).reset_index(drop = True)
similarities_df1
```
